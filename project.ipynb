## PART 1: DATA PREPERATION
### Loading the dataset

```python
# Import necessary libraries
import pandas as pd

# Load the dataset
heart_disease_df = pd.read_csv("heart_disease.csv")

# Display the first 50 rows of the dataframe
heart_disease_df.head(10)
```

## Inspecting the data to:
- Display information about the variables
- Display the data types of each column
- Display the unique values in each column

```python
# Check the dimensions of the dataframe
dimensions = heart_disease_df.shape
print("Dimensions of the dataframe:", dimensions)

# Display information about the variables, including data types
data_info = heart_disease_df.info()

# Display the unique values in each column
unique_values = heart_disease_df.nunique()

# Display the data types of each column
data_types = heart_disease_df.dtypes

# Print the results
print("\nInformation about the variables:")
print(data_info)

print("\nUnique values in each column:")
print(unique_values)

print("\nData types of each column:")
print(data_types)
```

## Types of Data
### Identifying two categorical variables and indicating if they are nominal or ordinal 

```python
# Check for 'Gender'
gender_column = 'Gender'
if gender_column in heart_disease_df.columns:
    unique_values = heart_disease_df[gender_column].unique()
    variable_type = 'Nominal'
    print(f"{gender_column}: {variable_type}")

# Check for 'education'
education_column = 'education'
if education_column in heart_disease_df.columns:
    unique_values = heart_disease_df[education_column].unique()
    variable_type = 'Ordinal'
    print(f"{education_column}: {variable_type}")
```

### Finding the missing values

```python
import pandas as pd

# Assuming 'heart_disease_df' is your dataframe
missing_values = heart_disease_df.isnull().sum()

missing_percentage = (missing_values / len(heart_disease_df)) * 100

missing_percentage = missing_percentage.round(2)

missing_info = pd.DataFrame({
    'Missing Values': missing_values,
    'Percentage': missing_percentage
})

print("Missing values information:")
print(missing_info)
```

### Updating the Rows with the Missing Values

```python
import pandas as pd

# Assuming 'heart_disease_df' is your dataframe
columns_with_missing_data = ['education', 'cigsPerDay', 'BPMeds', 'totChol', 'BMI', 'heartRate', 'glucose']

# Drop rows with missing values in the specified columns
heart_disease_df.dropna(subset=columns_with_missing_data, inplace=True)

# Display the updated dimensions of the DataFrame
updated_dimensions = heart_disease_df.shape
print(f"Updated Dimensions of the DataFrame: {updated_dimensions}")

# Check for missing values after the update
updated_missing_values = heart_disease_df.isnull().sum()
updated_missing_percentage = (updated_missing_values / len(heart_disease_df)) * 100

# Create a DataFrame with updated missing values information
updated_missing_info = pd.DataFrame({
    'Updated Missing Values': updated_missing_values,
    'Updated Percentage': updated_missing_percentage
})

# Print the updated missing values information
print("Updated Missing values information:")
print(updated_missing_info)
```

## converting categorical variables into numerical format

```python
from sklearn.preprocessing import LabelEncoder

# Assuming 'heart_disease_df' is your dataframe
label_encoder = LabelEncoder()

# Convert 'Gender' column to integers
heart_disease_df['Gender'] = label_encoder.fit_transform(heart_disease_df['Gender'])

# Convert 'education' column to integers
heart_disease_df['education'] = label_encoder.fit_transform(heart_disease_df['education'])

# Convert 'prevalentStroke' column to integers
heart_disease_df['prevalentStroke'] = label_encoder.fit_transform(heart_disease_df['prevalentStroke'])

# Convert 'Heart_stroke' column to integers
heart_disease_df['Heart_ stroke'] = label_encoder.fit_transform(heart_disease_df['Heart_ stroke'])

# Display the updated dataframe
print(heart_disease_df.head())
```
